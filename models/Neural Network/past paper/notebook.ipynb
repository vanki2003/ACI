```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')

pwd = os.getcwd()

df = pd.read_csv(pwd +'\shuttle_custom.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# dropping null values

df.dropna(inplace=True)
df.describe()
```

```python
# dropping duplicates

df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.corr().style.background_gradient(cmap='viridis')
```

```python
# visualizing the distribution of the target variable per class
df.columns

variables = ['A1', 'A2', 'A3', 'A4', 'A5', 'A6', 'A7', 'A8', 'A9']

for variable in variables:
    plt.figure(figsize=(10, 8))
    sns.histplot(x=variable, hue='target', data=df, multiple='stack')
    plt.show()

```

```python
target = df['target']

predictors = df.drop('target', axis=1)

predictors.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_df = scaler.fit_transform(predictors)

scaled_df
```

```python
from sklearn.model_selection import train_test_split

X = scaled_df
y = target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(X_train.shape)
print(X_train.shape)
```
